```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gaussian(x, sig):
    """
    mu is the mean ==0
    sig is standard devaiation
    """
    return np.exp(-np.power((x)/sig, 2.)/2)
```

```python
def hp(theta):
    '''
    theta is size in arcseconds
    hp is half power scale in kilolambda
    '''
    halfpower=92./theta # no idea where the 92 comes from
    return halfpower
```

```python
print hp(4.0),hp(7.0)
```

```python
def sigma(theta):
    s=hp(theta)*0.9
    return s
```

```python
print sigma(4.0)
print sigma(7.0)
```

```python

kl=np.linspace(0.,25,25)
half=0.5
plt.figure(figsize=(12,12))
plt.title('Gaussian sources')
plt.axhline(y=0.5, color="black", linestyle=":")
plt.axhline(y=0.9, color="red", linestyle="--")
plt.minorticks_on()
plt.plot(kl,gaussian(kl,sigma(1.)),label='1')
plt.plot(kl,gaussian(kl,sigma(2.)),label='2')
plt.plot(kl,gaussian(kl,sigma(3.)),label='3')
#
plt.plot(kl,gaussian(kl,sigma(4.)),label='4')
plt.plot(kl,gaussian(kl,sigma(5.)),label='5')
plt.plot(kl,gaussian(kl,sigma(6.)),label='6')
plt.plot(kl,gaussian(kl,sigma(7.)),label='7')
#plt.plot(kl,half)
plt.xlabel('$k\lambda$')
plt.ylabel('relative response')
plt.legend(title='size/arcsec',loc='center left')
plt.grid()
```
