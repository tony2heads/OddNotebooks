```python
from astropy.io import ascii
import astropy.coordinates as coord
import astropy.units as u
import numpy as np
```

```python
cals=ascii.read("/home/tony/Desktop/SouthAfrica/SourceLists/cals_Lband.csv")
names=cals['name']
rad=coord.Angle(cals['ra'],unit=u.hour).degree
decd=coord.Angle(cals['dec'], unit=u.degree).degree
print len(names), len(decd), names[0], decd[0:4]
```

```python
def ecliptic(r):
    ec=23.43487*np.sin(np.radians(r))
    return ec

def galaxy(r):
    #atan(A*sin((x-282.25)/57.29577)
    A=2.049016 # radians(180-62.6)
    g=np.degrees(np.arctan(A*np.sin(radians(r-282.25))))
    return g
    
```
