# w Term

baseline length B in wavelengths

$ B= \sqrt(u^2 +v^2 +w^2)$

* u,v are the well known u,v coordinates
* w is delay term

If we do this in direction cosines l,m,n


# Consider a non-zero area of sky with sources

* for the centre
$B^2={u_0}^2 +v{_0}^2 +w{_0}^2$ 

divide all by B and express as direction cosines l,m,n
where

$l = \frac{u}{B}$

$m =\frac{v}{B}$

$n=\frac{w}{B}$

$1=l^2+m^2+n^2$

## Consider two positions

$n_0 = \sqrt{1- ({l_0}^2 +{m_0}^2)} $


*for some other position __not far away__

$n_1= \sqrt{1- ({l_1}^2 +{m_1}^2)} $


* Difference

$n_1 -n _0 = \sqrt{1- ({l_0}^2 +{m_0}^2)} - \sqrt{1- ({l_1}^2 +{m_1}^2)} $


If I replace 
$ x = \sqrt{ l^2 + m2}$

__Note__ that x has a range [0,1]

* as x tends to zero, n tends to 1,so the source is high and delay is low
* as x tends to 1, n tends to 0, the source is low and delay is high (maximum at x=0)

$ \Delta n = n_0 - n_1  = \sqrt{1-x_1} - \sqrt{1- x_0}$
