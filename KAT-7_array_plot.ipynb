```python
# %load '/home/tony/Desktop/PythonScripts/pyx/KATposgrid.py'
#!/usr/bin/env python
%pylab inline

from pyx import *

unit.set(defaultunit='mm')

c=canvas.canvas()



#XYZ positions in metres from katfile ENU values
ants=(("1", 25.095, -9.095, 0.045),\
          ("2", 90.289, 26.377,-0.236),\
          ("3",  3.985, 26.909,-0.013),\
          ("4",-21.594, 25.508, 0.008),\
          ("5",-38.266, -2.565, 0.367),\
          ("6",-61.584, -79.648,  0.662),\
          ("7",-87.980, 75.778,  0.136))
for a in ants:
    name=a[0]
    #print name
    x,y=a[1],a[2]
    
    c.text(x,y,name,[text.size.LARGE,text.halign.center,text.valign.middle])
    c.stroke(path.circle(x,y,6)) # 6m radius
# add an arrow pointing north
length=40#metres
offx=40
offy=50
c.stroke(path.line(offx,offy,offx,length+offy),[deco.arrow(size=3),style.linewidth.THIck]) # North
c.stroke(path.line(offx,offy,length+offx,offy),[deco.arrow(size=3),style.linewidth.THIck]) #East

# line is from,to (x1,y1,x2,y2)
c.text(offx+10,(length/3+offy),"North ",[text.size.Large, trafo.rotate(90)]) #position

c.text((length/3+offx),offy+2,"East ",[text.size.Large]) #position

# Now a grid 

a=100 #m size of array
for x in range(-a,a,10): #10m steps
    c.stroke(path.line(x,-a,x,a), [style.linestyle.dashed,style.linewidth.thin])
    c.stroke(path.line(-a,x,a,x), [style.linestyle.dashed,style.linewidth.thin])
c.stroke(path.line(0,-a,0,a),[style.linewidth.THIck])
c.stroke(path.line(-a,0,a,0),[style.linewidth.THIck])



```

```python
c.writeEPSfile("KAT7arrayLayout")
#c.writePDFfile("KAT7arrayLayout")
```
