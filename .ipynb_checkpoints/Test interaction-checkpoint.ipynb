```python
%matplotlib notebook
import matplotlib.pyplot as plt
from ipywidgets import interact


fig, ax = plt.subplots()
ax.plot(range(5))


vline = ax.axvline(1, color='k')
hline = ax.axhline(0.5, color='k')

def set_cursor(x, y):
    vline.set_xdata((x, x))
    hline.set_ydata((y, y))
    ax.figure.canvas.draw_idle()
    plt.draw()
```

```python
interact(set_cursor,x=(-1,1,.1),y=(-1,1,.1))
```
