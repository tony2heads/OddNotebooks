# Spectrum fits

Several places (Perley & Butler, give log polynomials in GHz)

$Log(S) = A + B\times(log(\nu)) + C\times(log(\nu))^2 +D\times(log(\nu))^3$

But I want 

$Log(S) = a + b\times(log(F)) + c\times(log(F))^2 +d\times(log(F))^3$

Where F is in MHz

# recast

* a=A -3*B + 9*C  -27*D
* b=B -6*C +27*D
* c=C -9*D
* d=D

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def toM(A,B,C,D):
    a=A -3*B + 9*C  -27*D
    b=B -6*C +27*D
    c=C -9*D
    d=D
    return(a,b,c,d)
```

# First for 3C123

```python
def p123(f):# Perley
    n=f-3
    s= 1.8017 -0.7884*n  -0.1035*n**2 +0.0248*n**3 +0.0090*n**4
    return s
def b123(f): #Baars
    s=2.921-0.002*f -0.124*f**2
    return s
def o123(f): #Ott
    s=2.525+0.246*f -0.1638*f**2 #Ott
    return s
```

```python
x=np.linspace(2.,4.,99)
y=p123(x)
p=np.polyfit(x,y,3)
print p[::-1]

def fake123(f):
    a=p[3]
    b=p[2]
    c=p[1]
    d=p[0]
    print a,b,c,d
    s=a + b*f +c*f**2 +d*f**3
    return s 
```

```python
f=np.linspace(2.,4,40) # from 100MHz to 10GH
plt.figure(figsize=(10,10))
plt.plot(f,p123(f),label="3C123 Perley",color='r',ls="--")
plt.plot(f,b123(f),label="3C123 Baars")
plt.plot(f,o123(f),label="3C123 Ott",color='k')
plt.plot(f,fake123(f),label='approx',ls="-.",color='g')
plt.grid()
plt.legend()
plt.show()
```

```python
#Taurus

def pTau(f): #Perley
    n=f-3
    s= 2.9516 -0.217*n -0.047*n**2 -0.067*n**3  # Perley-butler 2016
    return s
def bTau(f): #Baars
    s=3.915 -0.299*f
    return s

    
```

```python
def fakeTau(f):
    A=2.9516
    B=-0.217
    C=-0.047
    D=-0.067    
    a=A -3*B +9*C -27*D
    b=B -6*C+27*D
    c=C -9*D
    d=D
    print a,b,c,d
    s=a + b*f +c*f**2 +d*f**3
    return s 
```

```python
f=np.linspace(2.,4,40) # 100MHz to 10GHz
plt.figure(figsize=(12,12))
plt.plot(f,pTau(f),label="Tau Perley",ls="-.",color='r')
plt.plot(f,bTau(f),label="Tau Baars")
plt.plot(f,fakeTau(f),label="reconst",ls=":")
plt.grid()
plt.legend()
plt.show()
```

```python
def pHyd(f): #Perley
    n=f-3
    s=1.7795 -0.9176*n -0.084*n**2 -0.0139 *n**3  +0.030*n**4  # Perley-butler 2016
    return s
def bHyd(f): #Baars
    s=4.497 -0.910*f
    return s
def oHyd(f): #Ott
    s=4.729 -1.025*f +0.0130*f**2
    return s

    
```

```python
x=np.linspace(2.,4.,99)
y=pHyd(x)
p=np.polyfit(x,y,3)
print p[::-1]

def fakeHyd(f):
    s=p[3] +p[2]*f +p[1]*f**2 +p[0]*f**3
    return s
```

```python
f=np.linspace(2.,4,40) # 100MHz to 10GHz
plt.figure(figsize=(12,12))
plt.plot(f,pHyd(f),label="Hydra Perley")
plt.plot(f,bHyd(f),label="Hydra Baars")
plt.plot(f,oHyd(f),label="Hydra Ott")
plt.plot(f,fakeHyd(f),label="Hydra approx")
plt.grid()
plt.legend()
plt.show()
```
