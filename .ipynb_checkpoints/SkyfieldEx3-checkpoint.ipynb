```python
import matplotlib.pyplot as plt
import numpy as np
from skyfield.api import load

data  = load('de421.bsp')
earth = data['earth']
moon  = data['moon']
bary  = data['earth barycenter']

ts=load.timescale(builtin=True)
day0=ts.utc(2020, 1, 1, 12, 0, 0)
print day0
days = range(0, 366, 5)
earthpos, moonpos, barypos = [], [], []
for day in days:
    date=ts.utc(2020,day)
    earthpos.append(earth.at(date).position.km)
    moonpos.append(moon.at(date).position.km)
    barypos.append(bary.at(date).position.km)
ep = np.array(earthpos).T
mp = np.array(moonpos).T
bp = np.array(barypos).T

plt.figure(figsize=[9,9])
plt.subplot(5,1,1)
for thing in ep:
    plt.plot(days, thing)
plt.subplot(5,1,2)
for thing in mp:
    plt.plot(days, thing)
plt.subplot(5,1,3)
for thing in bp:
    plt.plot(days, thing)
plt.subplot(5,1,4)
for thing in (ep-bp):
    plt.plot(days, thing)
plt.subplot(5,1,5)
for thing in (mp-bp):
    plt.plot(days, thing)
plt.savefig('bary two')
plt.show()
```
