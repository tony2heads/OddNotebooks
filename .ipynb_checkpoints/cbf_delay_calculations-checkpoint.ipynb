```python
data = '/var/kat/archive2/data/MeerKATAR1/telescope_products/2017/03/29/1490791378.h5'

f = katdal.open(data)
```

```python
array_ant = katpoint.Antenna('ref', *(f.ants[0].ref_position_wgs84))
correlator_model = katpoint.DelayCorrection(f.ants, array_ant)
```

```python
targ = f.catalogue.targets[0]
delaylist = correlator_model.corrections(targ,f.timestamps)
```
