# Analytic approximation for SEFD
## from Marisa's Giant Pulses paper
* frequency in MHz
* SEFD is in K/Jy
For a single dish

```python
def SEFD(f):
    """
    freuqncy f in MHz
    SEFD in K/Jy
    """
    s=5.71e-7*f**3  -1.90e-3*f**2 +1.90*f -1.13e2
    return(s)
    
```

```python
print(SEFD(1220.4))
```

```python
import matplotlib.pyplot as pl
%matplotlib notebook
```

```python
import numpy as np
```

```python
f=np.linspace(881.0,1670.0,88) # stable range is 881 to 1670
pl.plot(f,SEFD(f))
pl.grid()
pl.xlabel("frequency MHz")
pl.ylabel("SEFD K/Jy")
pl.title("Analytic approximation for SEFD from Giant Pulses paper")
```
