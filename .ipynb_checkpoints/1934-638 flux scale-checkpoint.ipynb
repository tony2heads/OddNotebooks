```python
import numpy as np
%pylab inline
```

```python
def reynolds(lf):   #Also in Miriad calstoke.for
    a=-30.7667
    b=26.4908
    c=-7.0977
    d=0.605334
    return a+b*lf +c*lf*lf + d*lf*lf*lf
```

```python
def sault(lf): # only valid above 10GHz
    a=-202.6259
    b=149.7321
    c=-36.4943
    d=2.9372
    return a+b*lf +c*lf*lf + d*lf*lf*lf
```

```python
def old(lf):
    a=-31.0214
    b=26.7694
    c=-7.1947
    d=0.6162
    return a+b*lf +c*lf*lf + d*lf*lf*lf
```

```python
def old2(lf): # in calstoke.for
    a=-23.839
    b=19.569
    c=-4.8168
    d=0.35836
    return a+b*lf +c*lf*lf + d*lf*lf*lf
```

```python
def my(lf): #attempted fit for 408MHz-20GHz, ignoring PAPER data point
    a=-30.1187
    b=25.8040
    c=-6.8601
    d=0.5787
    return a+b*lf +c*lf*lf + d*lf*lf*lf
```

```python
low=np.log10(856)
high=np.log10(1712)
l=np.linspace(low,high)
pylab.figure(figsize=(20,20))
pylab.plot(10**l,10**reynolds(l),label="Reynolds")
pylab.plot(10**l,10**old(l),label="Old")
pylab.plot(10**l,10**old2(l),label="Old2")
#pylab.plot(10**l,10**sault(l),label="Sault")
pylab.plot(10**l,10**my(l),label="mine")
pylab.grid()
pylab.legend()
```
