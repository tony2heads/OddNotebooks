```python
import numpy as np
%pylab inline
```

```python
def fwhmH(fr):
    """
    freq in MHz
    wavelength in meters
    """
    wavel=2.998e2/fr
    w=1.028*wavel/12.2 #radians
    arcmin=np.degrees(w)*60.
    return arcmin
def fwhmV(fr):
    wavel=2.998e2/fr
    w=1.028*wavel/11.8
    arcmin=np.degrees(w)*60.
    return arcmin
```

```python
def cos2(r,theta):
    t=theta/60.0 #min to deg
    # r in degrees theta in arcmin
    x=1.189*r/t
    #print x
    a=(np.cos(x*np.pi)/(1 - 4*x**2))**2
    return a
```

```python
def db(y):
    return 10*np.log10(y)
```

```python
f= 1000. #MHz
theta= fwhmH(f)
print("FWHM",theta/60. ,"Degrees")
print("HWHM",theta/120. ,"Degrees")
```

```python
figsize(25,25)
```

```python
r=np.linspace(0,2.5,301)
rc('font',size=18)
fs=22 # font size
title("Mauch et. al. model",fontsize=fs)
plot(r,db(cos2(r,theta)),label="response")
#ylim(0.0,1.0)
xlabel("Distance/deg",fontsize=fs)
ylabel("power response/dB",fontsize=fs)
grid('on','both')
```

```python
max_off=2.5 # range in degrees 
ticks = np.arange(-max_off, max_off+0.5, 0.5)
yticks=np.arange(0,1,0.1)
r=np.linspace(-max_off,max_off,301)
rc('font',size=18)
fs=22 # font size
axes(xticks=ticks,yticks=yticks)
title("Mauch et. al. model Lband",fontsize=fs)
plot(r,cos2(r,theta),label="relative power "+str(int(f))+"MHz")
#plot(r,0.5,label="half power")
legend()
ylim(0.0,1.0)
xlim(-max_off,max_off)
xlabel("Distance/deg",fontsize=fs)
ylabel("linear power response",fontsize=fs)
hwhm= fwhmH(f)/120
x1=hwhm
x2=-1*x1
annotate("half power",(x1,0.5))
annotate("half power",(x2,0.5))
vlines(x1,0,1)
vlines(x2,0,1)
#plot(x2,0.5)
grid('on','both')
savefig("Antenna response"+str(int(f))+"MHz.png")
```
