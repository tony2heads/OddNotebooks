```python
import matplotlib.pyplot as plt
from skyfield.api import load, JulianDate

ts = load.timescale()
t = ts.utc(1980, 4, 20)
#t = ts.tt(jd=2444349.500592
data = load('de421.bsp')
sun  = data['sun']
bary = data[10]

years = [1975+i for i in range(51)]
sunpos, barypos = [], []

for year in years:
    jd = ts.utc(year, 1, 1)
    sunpos.append(sun.at(jd).position.km)
    barypos.append(bary.at(jd).position.km)

plt.figure()
x, y, z = zip(*sunpos)
plt.plot(years, x)
plt.plot(years, y)
plt.plot(years, z)
# x, y, z = zip(*barypos)
# plt.plot(years, x, '-k')
# plt.plot(years, y, '-k')
# plt.plot(years, z, '-k')

plt.title('suns motion in barycentric frame')
plt.savefig('bary one')
plt.show()
```
