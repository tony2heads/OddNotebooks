```python
from pygsm import GlobalSkyModel
from pygsm import GlobalSkyModel2016
from pygsm import GSMObserver
from datetime import datetime
from pygsm import GSMObserver2016
```

```python
# Setup observatory location - in this case, MeerKAT
(latitude, longitude, elevation) = ('-30.72148', '21.388', 1034)
ov = GSMObserver()
ov.lon = longitude
ov.lat = latitude
ov.elev = elevation
ov.date = datetime(2018, 2, 20, 6, 30)

freq=1500.
```

```python
ov.generate(freq)
d = ov.view(logged=True)
```

```python
ov = GSMObserver2016()
ov.lon = longitude
ov.lat = latitude
ov.elev = elevation
ov.date = datetime(2018, 2, 20, 6, 30)
ov.generate(freq)
d = ov.view(logged=True)
```

```python
d = ov.view_observed_gsm()
```

```python

# Compare to 2008 De-Olivera Costa version
gsm_2008 = GlobalSkyModel(freq_unit='MHz',interpolation='pchip',basemap='haslam')
# can have  'haslam' 5deg' or 'wmap'
gsm_2008.generate(freq)
gsm_2008.view(logged=True)
```

```python
gsm_2016 = GlobalSkyModel2016(freq_unit='MHz',resolution='hi',unit='TCMB') # TRJ or TCMB or MJysr
```

```python

gsm_2016.generate(freq)
gsm_2016.view(logged=True)
```
