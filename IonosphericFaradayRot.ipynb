```python
def pha(FR,f):
    c=2.998e8 # m/s
    l=c/f
    r=FR*l**2
    p=np.degrees(r)
    return p
```

```python
f=np.linspace(2e8,2e9,181)
#print f
FR=2.0
#print pha(FR,f)
```

```python
pylab.figure(figsize=[15,15])
pylab.plot(f,pha(FR,f))
pylab.ylim(0.,45.)
pylab.xlim(5.8e8,1.7e9)
pylab.xlabel("Freq/Hz")
pylab.ylabel("Rotation/deg")
pylab.title("Slant RM=2.5 TECU~40")
pylab.grid()
pylab.show()
```
