```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gaussian(x, mu, sig):
    """
    mu is the mean
    sig is standard devaiation
    """
    return np.exp(-np.power((x - mu)/sig, 2.)/2)
```

```python
x=np.linspace(-3,3,50)
amp=gaussian(x,0,1.0)
```

```python
plt.figure(figsize=(10,10))
plt.minorticks_on()
plt.plot(x,amp)
plt.grid()
```

```python
def where(amp):
    """
    valid in the range 0 to 1.
    """
    x=np.sqrt(2*np.log(1/amp))
    return x
```

```python
print where(0.5), where(0.9)
```

```python
fwhm=2*where(0.5)
print "FWHM =twice HWHM = ",fwhm , "in units of Sigma"
```
