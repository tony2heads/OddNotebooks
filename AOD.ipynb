```python
import pysolr
from IPython.core.display import display, HTML
archive = pysolr.Solr('http://192.168.1.50:8983/solr/kat_core')
search_str = []
search_str.append('CAS.ProductTypeName:MeerKATAR1TelescopeProduct')
search_str.append('StartTime:[2017-03-07T00:00:00Z TO 2017-03-15T00:00:00Z]')
search_string = ' AND '.join(search_str)
ar_res = archive.search( search_string ,rows=10000,**{'sort':'StartTime desc'})
h = '<table>'
h += "<tr><th>Owner<th>UT Start<th>Schedule block<th>Description<th>chan<th>ants<th>size<th>Archive</tr>"			
for r in ar_res.docs[::-1]:
    link = 'http://kat-archive.kat.ac.za:8080/archive_search/#{"ei":"%s","so":true,"r":true,"k":true,"a":true,"b":true,"ob":"1","rs":{}}'%(r['ScheduleBlockIdCode'])		
    h += '<tr> <td> %s <td> %s <td> <a href=\'%s\'>%s</a> <td> %s <td> %d <td> %d <td> %5.2f <td> %s  </tr>'%(r['Observer'],r['StartTime'],link,r['ScheduleBlockIdCode'],r['Description'], r['NumFreqChannels'],len(r['Antennas']),r['FileSize'][0]/1024.0**3,r['CAS.ReferenceDatastore'][0][5:])		
h += "</table>"
display(HTML(h))
h += "<tr><th>Owner<th>UT Start<th>Schedule block<th>Description<th>chan<th>ants<th>size<th>Archive</tr>"			
from BeautifulSoup import BeautifulSoup
soup = BeautifulSoup(h)
with open("output.html", "wb") as file:
    file.write("<html>\n")
    file.write(soup.prettify("utf-8"))
    file.write("\n<\html>\n")
```
