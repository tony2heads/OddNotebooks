```python
for filename in filenames:
    print filename
    h = katdal.open(filename)
    k = h.file['TelescopeState']['cal_product_K'] # Delays
    k_list = []
    for i, ki in enumerate(k):
        k_list.append(pickle.loads(ki[1]))
    k_array = np.array(k_list)
    k_0 = k_array[:,0,:][-1]*1e9 # scale from seconds to nanoseconds Last value ONLY
    k_1 = k_array[:,1,:][-1]*1e9
    #
    g = h.file['TelescopeState']['cal_product_G'] # Complex gain values
    g_list = []
    for i, gi in enumerate(g):
        g_list.append(pickle.loads(gi[1]))
    g_array = np.array(g_list)
    #print g_array.shape
    g_0 = g_array[:,0,:]
    g_1 = g_array[:,1,:]
    dH=np.vstack((DelayH,k_0))
    dV=np.vstack((DelayV,k_1))
    DelayH=dH # update with new vals
    DelayV=dV
    aHS=np.vstack((AmpH_Start,np.abs(g_0)[0]))
    aHE=np.vstack((AmpH_End,np.abs(g_0)[-1]))
    AmpH_Start=aHS # update
    AmpH_End=aHE
    aVS=np.vstack((AmpV_Start,np.abs(g_1)[-1]))
    aVE=np.vstack((AmpV_End,np.abs(g_1)[-1]))
    AmpV_Start=aVS # update
    AmpV_End=aVE
    pHS=np.vstack((PhaseH_Start,np.rad2deg(np.angle(g_0)[0])))
    pHE=np.vstack((PhaseH_End,np.rad2deg(np.angle(g_0)[-1])))
    PhaseH_Start=pHS # update
    PhaseH_End=pHE
    pVS=np.vstack((PhaseV_Start,np.rad2deg(np.angle(g_1)[0])))
    pVE=np.vstack((PhaseV_End,np.rad2deg(np.angle(g_1)[-1])))
    PhaseV_Start=pVS #update
    PhaseV_End=pVE
```
