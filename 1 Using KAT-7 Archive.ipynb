This same method of search can be used in the command line interface from IPython.

```python
import katarchive
```

```python
fn = katarchive.search_archive(description='*health*')
print fn
```

If the filename is known, the archive can be queried directly for the file

```python
fn = katarchive.search_archive(filename='1362743782.h5')
print fn
```

The observation file can then be read and processed

```python
import katfile
```

```python
h5 = katfile.open(fn[0].path_to_file)
print h5
```
