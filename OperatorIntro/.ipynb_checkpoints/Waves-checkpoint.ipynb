# Electromagnetic Waves

Waves have a frequency $f$ an amplitude (peak-to-zero height) and a phase offset
$\phi$

What we measure at time $t$ is
$$ A=amp\times sin(2{\pi}ft +\phi)$$

Period of wave $T = 1/f$

```python
import numpy as np
import pylab as pl
%pylab inline
```

```python
freq=0.11 # picked for convenience
t=np.linspace(0,50,400)  # a range of time 0 to 50 in 400steps
amp=1.0
phase=0.0 #phase is zero at start
pl.figure(figsize=(14,4))
pl.plot(t,amp*np.sin(2*np.pi*freq*t+phase))
pl.show()
```

```python

phase1=1.5 #phase in radians at zero at start
pl.figure(figsize=(14,4))
pl.plot(t,amp*np.sin(2*np.pi*freq*t+phase1))
pl.show()
```

# So with two at different phases

```python

phase1=0.0 #phase in radians at zero at start
phase2=2.7
pl.figure(figsize=(14,4))
pl.plot(t,amp*np.sin(2*np.pi*freq*t+phase1))
pl.plot(t,amp*np.sin(2*np.pi*freq*t+phase2))
pl.show()
```

# What if you add them

```python
wavesum=(amp*np.sin(2*np.pi*freq*t+phase1) + amp*np.sin(2*np.pi*freq*t+phase2))

pl.figure(figsize=(14,4))
txt="Phases differ by "+str(phase2-phase1)
pl.plot(t,wavesum,label=txt)
pl.legend()
pl.show()
```
