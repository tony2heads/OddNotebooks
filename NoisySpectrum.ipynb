```python
"""
Get 2 sinusoid waves, add noise - plot
Do FFT  - plot
"""
from scipy import sin,cos,fft,ifft
#from pylab import *
%pylab inline
```

```python
t=arange(-50,50,0.01) # time axis
s=sin(2*2*pi*t)
c=cos(3*2*pi*t)

noise=10*randn(len(t)) # gaussian random noise scaled
signal=s+c+noise
# plot it

pylab.figure(1,figsize(12,12))
plot(t,noise,label="Gaussian noise")
plot(t,signal,label="sum")
plot (t,s,label="$sin(2\pi 2t)$") #nice greek letter
plot (t,c,label="$cos(2\pi 3t)$")
legend()
pylab.show()
```

```python
#pylab.figure(2)
scale=30./len(t)
print ifft(signal)
pylab.plot(t,scale*real(fftshift(ifft(signal))),label='real fft')
pylab.plot(t,scale*imag(fftshift(ifft(signal))),label='imag fft')
pylab.xlim(xmin=0,xmax=10) # only positive axis
pylab.legend()
pylab.show()
```
