```python
"""
Some exmples with matplotlib pylab
and scipy of Fourier Transforms
default order of the fast fourier transform is
0...n/2,-1,...(n/2)-1
so it needs reodering

"""
%pylab inline
from scipy import *
```

```python
def tophat(x,w):
    """
    tophat of total width w
    """
    d=w/2.
    f=zeros(len(x))
    for i in range(len(x)):
        if abs(x[i]) > d:
            f[i]=0
        else:
            f[i]=1
    return(f)
```

```python
def triangle(x,w):
    """
    triangle of total width w
    """
    d=w/2.
    f=zeros(len(x))
    for i in range(len(x)):
        if abs(x[i]) > d:
            f[i] = 0
        else:
            if x[i] < 0 :
                f[i]= 1 + x[i]/d
            if x[i] > 0 :
                f[i]= 1 - x[i]/d
    return f
```

```python
x=arange(-10*pi,10*pi,0.01)
xmin=min(x)
xmax=max(x)

#

scale=2./len(x)

f=cos(x)

figure(1)
plot (x,scale*real(fftshift(fft(f))),label='real fft')
plot (x,scale*imag(fftshift(fft(f))),label='imag fft')
plot (x,f,label='cos')
x=arange(-10*pi,10*pi,0.01)
xmin=min(x)
xmax=max(x)
legend()

#
```

```python
width=5.
f=tophat(x,width)

figure(2)
plot (x,width*scale*real(fftshift(fft(f))),label='real fft')
plot (x,width*scale*imag(fftshift(fft(f))),label='imag fft')
plot (x,f,label='tophat')
legend()
```

```python
width=4.
f=triangle(x,width)

figure(3)
plot (x,width*scale*real(fftshift(fft(f))),label='real fft')
plot (x,width*scale*imag(fftshift(fft(f))),label='imag fft')
plot (x,f,label='triangle')
legend()
```

```python
width=2.5
f=cos(x*pi/(width))*tophat(x,width)

figure(4)
plot (x,width*scale*real(fftshift(fft(f))),label='real fft')
plot (x,width*scale*imag(fftshift(fft(f))),label='imag fft')
plot (x,f,label='coshat')
legend()
show()
```
