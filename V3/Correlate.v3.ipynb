# Correlation functions

```python
import numpy as np
def noise(T):
    sigma=np.sqrt(T) # ignore Boltzmann's constant
    x=np.random.normal(0,sigma,10000000) # give me lots of random values with normal distribution
    return x
ant1=noise(30)


%pylab inline
sigma =np.std(ant1, ddof=1)

count, bins, ignored = pylab.hist(ant1, 30, normed=True)
pylab.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins)**2 / (2 * sigma**2) ),linewidth=2, color='r')
pylab.show()
```

```python
pylab.plot(ant1[0:10000])
pylab.show()
print # its just lots of noise
```

```python
ant2=noise(30)
signal=noise(1)
```

```python
pylab.plot(signal[0:10000])
pylab.show()

# this is just noise as well - but weaker
```

```python
pylab.plot(ant1[0:10000]+signal[0:10000],color='black')
pylab.plot(ant1[0:10000],color='red', alpha=0.5)
pylab.plot(ant2[0:10000],color='blue',alpha=0.25)

pylab.show()
#I  can't see it
```

```python
count, bins, ignored = pylab.hist(ant2+signal, 30, normed=True)
pylab.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins)**2 / (2 * sigma**2) ),linewidth=2, color='r')
pylab.show()
```

```python
def correlate(x,y,n=10000):
    prod=x*y        # mutliply two signals
    av=numpy.mean(prod.reshape(-1, n), 1) # average up by n; n must divide into the array length for our case
    return av

p=correlate(ant1,ant2,n=10)
print len(p)
```

```python
#what does cross correlation look like
pylab.plot(correlate(ant1,ant1)) # an autocorrelation
pylab.plot(correlate(ant2,ant2)) # another autocorrelation
pylab.plot(correlate(ant1,ant2)) # a cross correlation but no external signal
pylab.grid()
pylab.show()
```

# What about adding in a signal from the source 

```python
s1=ant1+signal
s2=ant2+signal
# same weak signal in BOTH
```

```python
pylab.plot(correlate(s1,s1))
pylab.plot(correlate(s2,s2))
pylab.plot(correlate(s1,s2))
pylab.grid()
pylab.show()
```
