```python
import pylab
from scipy import *
"""
This is the frequency version of blackbody enery density
"""

h=6.6244e-34 # Planck constant in Js
c=2.998e8     # speed of light in m/s
k=1.38e-23    #Boltzmann constant in J/K%

def blackbody(nu,T): # nu is frequency (Hz) , T is temperature (K)
    u_c=(8*pi*nu**2*k*T)/c**3 #classical part of energy density
    x=(h*nu)/(k*T)
    eb=x/(exp(x)-1 ) #Einstein-Bose correction factor - no units
    u_q=u_c*eb  # energy density
    S=u_q*c/4.0 # radiation
    return S
```

```python
"""
 NOTE: for radiation per unit are, multiply u by c/4
- factor of 1/2 from radiation going in both directions
- factor of 1/2 from a cos(theta) factor intgrated over angle in 2D
"""
%pylab inline
for T in (3,100,300,1000,3000,5000,10000): #temperatures
    lab=str(T)+'K'
    lognu=linspace(6.,14.,140) #140 equally spaced points
    nu=10**lognu
    pylab.ylim([-50,0])
    lb=log10(blackbody(nu,T))
    pylab.plot(lognu,lb,label=lab)
    pylab.xlabel('log freq/Hz')
    pylab.ylabel('log flux per unit frequency emitted')
    pylab.legend(loc='upper left')
    pylab.title("Blackbody radiation")
    pylab.ylim(-30,-5)
pylab.show()
```
