### Goodness of fit, R^2, sums of squares

A standard goodness of fit measure is the $R^2$ which
describes what percentage of the total variability of the 
outcome is described by the model in question. 
The $R^2$ is defined in terms 
of ratios of sums of squares. In the simple
linear regression case, we can explicitly visualize
these sums of squares.

These three figures illustrate the different 
sums of squares in the simple linear regression model.

The total sum of squares, $SST$, is the sum
of the squared differences between the *Y* 
values and the sample mean of the *Y* values.

```python
%load_ext course_modules.knitr_extension
```

```python
%%knitr
X = seq(0,20, length=21)
Y = 0.5*X+1 + rnorm(21)

Y.lm = lm(Y~X)

p = predict(Y.lm)
m = mean(Y)
plot(X, Y, pch=23, bg='red')
abline(h=m, col='yellow', lwd=2)
for (i in 1:21) {
  points(X[i], m, pch=23, bg='yellow')
  lines(c(X[i], X[i]), c(Y[i], m))
}
abline(Y.lm, col='green', lwd=2)
```

The error sum of squares, $SSE$, is the sum of the 
squared differences between the *Y* values and 
the *Yhat* values, i.e. the fitted values of the 
regression model.


```python
%%knitr

plot(X, Y, pch=23, bg='red')
abline(Y.lm, col='green', lwd=2)
for (i in 1:21) {
  points(X[i], p[i], pch=23, bg='green')
  lines(c(X[i], X[i]), c(Y[i], p[i]))
}

m = mean(Y)
abline(h=m, col='yellow', lwd=2)
```

The regression sum of squares, $SSR$, is the sum of the 
squared differences between the $\hat{Y}$ values
and the sample mean of the $Y$ values. 


```python
%%knitr

plot(X, Y, pch=23, bg='red')
abline(Y.lm, col='green', lwd=2)
abline(h=m, col='yellow', lwd=2)
for (i in 1:21) {
  points(X[i], p[i], pch=23, bg='green')
  points(X[i], m, pch=23, bg='yellow')
  lines(c(X[i], X[i]), c(m, p[i]))
}

```

### Definition of R^2

If the regression model fits very well, this sum
of squares will be large relative to the
total sum of squares. The $R^2$ is just
the ratio of these sums of squares.
$$
 R^2 = \frac{SSR}{SST}
$$

The $R^2$ is also closely related to the
$F$ statistic reported as the goodness
of fit in *lm*.

```python
%%knitr
url = 'http://stats191.stanford.edu/data/wage.csv'
wages = read.table(url, sep=',', header=T)
attach(wages)
wages.lm = lm(logwage ~ education)
```

The sums of squares are:

```python
%%knitr
SSE = sum(resid(wages.lm)^2)
SST = sum((logwage - mean(logwage))^2)
SSR = sum((mean(logwage) - predict(wages.lm))^2)
data.frame(SST, SSE+SSR)
```

```python
%%knitr
F = (SSR / 1) / (SSE/wages.lm$df)
F
sqrt(F)
summary(wages.lm)
```

### Poorly fitting linear model

Plots of residuals against covariates can sometimes
reveal missing nonlinear terms in a regression model.
In this example, the true mean is quadratic
and we try to use some diagnostic plots based
on residuals to "discover" this.


```python
%%knitr
url = "http://stats191.stanford.edu/data/anscombe.table"
anscombe <- read.table(url, header=T)
anscombe$Y2 = anscombe$Y2 + rnorm(length(anscombe$Y2)) * 0.45

plot(anscombe$X2, anscombe$Y2, pch=23, bg='orange', cex=2, ylab='Y', xlab='X')
simple.lm <- lm(anscombe$Y2 ~ anscombe$X2)
abline(simple.lm$coef, lwd=2, col='red', lty=2)
```

If we plot the residuals against *X*, we see a clear pattern
(in fact, a quadratic). Patterns in these residual plots
may suggest something like a quadratic effect is missing, but
they can also suggest some sort of serial dependence in the
random errors. We will discuss this later, when we discuss
*correlated-errors*.

```python
%%knitr
plot(anscombe$X2, resid(simple.lm), ylab='Residual', xlab='X',
     pch=23, bg='orange', cex=2)
abline(h=0, lwd=2, col='red', lty=2)
```

Next, we add a quadratic term to the model, using the function
*poly* which adds a polynomial trend of a given degree to a model.


```python
%%knitr
quadratic.lm <- lm(Y2 ~ poly(X2, 2), data=anscombe)
Xsort <- sort(anscombe$X2)
plot(anscombe$X2, anscombe$Y2, pch=23, bg='orange', cex=2, ylab='Y', xlab='X')
lines(Xsort, predict(quadratic.lm, list(X2=Xsort)), col='red', lty=2, lwd=2)
```


The residuals of the quadratic model have no apparent pattern in them,
suggesting this is a better fit than the simple linear regression model.


```python
%%knitr
plot(anscombe$X2, resid(quadratic.lm), ylab='Residual',
     xlab='X', pch=23, bg='orange', cex=2)
abline(h=0, lwd=2, col='red', lty=2)
```

Another common diagnostic plot is the *qqplot* where *qq* stands for 
*Quantile-Quantile*. Roughly speaking, a *qqplot* tries to see 
if the quantiles of two distributions match. The function *qqnorm* 
can be used to ascertain if a sample of numbers are roughly normally
distributed. If the points lie on the diagonal line, this is 
evidence that the sample is normally distributed. Various departures
from the diagonal indicate skewness, asymmetry, etc.


```python
%%knitr
qqnorm(resid(simple.lm), pch=23, bg='orange', cex=2)
qqnorm(resid(quadratic.lm), pch=23, bg='orange', cex=2)
```

In these two examples, the qqplot does not seem vastly different, even
though we know the simple model is incorrect in this case. This indicates
that several diagnostic tools should be used in assessing a model.


### Outliers and nonconstant variance

One of our assumptions is that the variance is constant, that
is, the spread of the residuals around the fitted line is roughly
the same at each value of the covariate. This dataset below, 
taken from some work done with Dr. Robert Shafer here at Stanford
http://hivdb.stanford.edu,
plots HIV virus load against a score related to the
the genetic makeup of a patient's virus shows clear non-constant variance.
It also provides a clear example of an outlier, or a 
point that is a clear departure from the model.


```python
%%knitr
url = 'http://stats191.stanford.edu/data/HIV.VL.table'
viral.load = read.table(url, header=TRUE)
attach(viral.load)

plot(GSS, VL, pch=23, bg='orange', cex=2)
viral.lm = lm(VL ~ GSS)
abline(viral.lm, col='red', lwd=2)
```

Let's refit, removing the outlier

```python
%%knitr
good = (VL < 250000)
plot(GSS, VL, pch=23, bg='orange', cex=2)
viral.lm.new = lm(VL ~ GSS, subset=good)
abline(viral.lm.new, col='green', lwd=2)
abline(viral.lm, col='red', lwd=2)
```

Plotting residuals against GSS shows a clear "fan" pattern,
which suggests that the variance of the errors grows
with GSS. This can sometimes be modelled, and corrected
for, using weighted regression.


```python
%%knitr
plot(GSS[good], resid(viral.lm.new), pch=23, bg='orange', cex=2, xlab='GSS', ylab='Residual')
abline(h=0, lwd=2, col='red', lty=2)
```
