```python
import pysolr
from IPython.core.display import display, HTML


startday="2017-09-17"
stopday="2017-09-24"

archive = pysolr.Solr('http://192.168.1.50:8983/solr/kat_core')

search_str = 'CAS.ProductTypeName:MeerKATAR1TelescopeProduct AND '
search_str += 'StartTime:['
search_str += startday
search_str += 'T00:00:00Z TO '
search_str += stopday
search_str += 'T23:59:59Z]'
ar_res = archive.search( search_str ,rows=10000,**{'sort':'StartTime desc'})
```

```python
from openpyxl import Workbook
wb = Workbook() # setup an excel workbook
ws=wb.active # one sheet
ws.title=startday+" to "+stopday
```

```python
for r in ar_res.docs[::-1]:
    if 'ScheduleBlockIdCode' not in r:
        r['ScheduleBlockIdCode']=""
    if len(r['Description'])<1 :
        r['Description']='Beamformer?'
    #print r['ScheduleBlockIdCode']
    id=r['ScheduleBlockIdCode']
    
    link = 'http://kat-archive.kat.ac.za:8080/archive_search/#{'
    link+='"s"":""'+id+'"",'
    link+='""a"":true,'
    link+='""rs"":{}}'

    print link
    gigs=r['FileSize'][0]/1024.0**3
    place=r['CAS.ReferenceDatastore'][0][5:]
    nants=len(r['Antennas'])
    id=r['ScheduleBlockIdCode']
    text='%s|%s|%s|%s|%s|%d|%d|%5.2f|%s' \
    %(r['Observer'],r['StartTime'],link,id,r['Description'],\
      r['NumFreqChannels'],nants,gigs,place)
    #print text
    ws.append(text.split("|")) 
```

```python

wb.save("AOD_"+startday+'.xlsx')
```
