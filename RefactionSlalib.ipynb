```python
zobs=1.0 # zenith distance in radians AFTER REFRACTION
hm=1038. #  height above sea level in metres
t= 300. # ambient temperature in Kelvin 
p= 900.0 # pressure in mbar
rh=0.3 # relative humidiity (0-1 scale)
wl=10000 # wavelength (>100 is radio)
ph=-0.5361909 # latitude in radian
tlr=0.0065 # lapse rate in K/m 
eps=1e-8 # precision
```

```python
ref=s.sla_refro(zobs,hm,t,p,rh,wl,ph,tlr,eps) #refraction coefficient -1
```

# from vacuum to observed

```python
def shiftr(zr,a,b):
    """
    Zenith angle REAL in radians 
    to refraction in arcsec
    """
    zout=s.sla_refz(zr,a,b)
    ref=3600.*np.rad2deg(zr-zout)
    return ref
print shiftr(np.deg2rad(80),a,b)
```

# from observed to vacuum

```python
def dz(zobs,a,b):
    #zobs in radian
    t=np.tan(zobs)
    deltaZ=a*t + b*t**3
    deltaZ=deltaZ*206265. # arcsec
    return deltaZ
```

```python

tests=80

vec_shift=np.vectorize(shiftr)
zobs=np.linspace(0.,np.deg2rad(tests),81)
#print testz,zobs, dz(zobs,a,b),vec_shift(zobs,a,b)
diff=(vec_shift(zobs,a,b)- dz(zobs,a,b))
print ','.join(['%-6.3f' % (i,) for i in diff])
```

```python
%pylab inline
zen=np.linspace(0,80.0,100)
z=np.deg2rad(zen)
#print z[-1], a,b
#
pylab.figure(figsize=(15.,15.))
pylab.plot(zen,dz(z,a,b) ,label="Slalib obs to real")
pylab.plot(zen,vec_shift(z,a,b),label='Slalib real to obs')
pylab.xlabel("ZA/deg")
pylab.ylabel("ref/arcsec")
pylab.grid()
pylab.legend()
pylab.show()
```
