# What pointing accuracy could we expect for L-band and Ku-band with MeerKAT

 - see Ivison R. J. et al., 2007, MNRAS, 380, 199
 
 $\sigma_{theta} \approx 0.6 \frac{ \theta_{FWHM}}{SNR}$


#L -band

$\theta_{FWHM}  \approx {1.1^{\circ}} =66 arcmin$

Typical source is 7Jy
Noise is from rms variation from the signal integrated over the bandwidth and time
confusion in a beam about 0.2Jy

```python
import math
Tsys=20  #K
gain=0.03 #K/Jy
BW=8e8 #Hz
t=1 #s
noise=(Tsys/gain)/math.sqrt(BW*t)
conf=0.2 #Jy
noise=math.sqrt(noise*noise + conf*conf)
signal = 7 #Jy
snr=signal/noise
theta=66 #arcmin
sigma=0.6*theta/snr
print("%3.1f arcmin") %( sigma)
```

# Ku band
$\theta_{FWHM}  \approx {0.12^{\circ}} =7 arcmin$

Typical source is 2Jy
Noise is from rms variation from the signal integrated over the bandwidth and time
confusion in a beam about 0.01Jy


```python
Tsys=100  #K
gain=0.03 #K/Jy
BW=8e8 #Hz
t=1 #s
noise=(Tsys/gain)/math.sqrt(BW*t)
conf=0.01 #Jy
noise=math.sqrt(noise*noise + conf*conf)
signal = 2 #Jy
snr=signal/noise
theta= 7.2 #arcmin
sigma=0.6*theta/snr
print("%3.1f arcmin") %( sigma)
```
